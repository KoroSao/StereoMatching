```python
def disparity_to_gray(disp, bgr=True):
    image = np.zeros((disp.shape[0], disp.shape[1]), dtype=np.uint8)
    is_occluded = disp < 0
    image[:,:] = np.where(is_occluded, 0, 255 * disp / disp.max())
    image[is_occluded] = [255, 255, 0] if bgr else [0, 255, 255]
    return image
```
